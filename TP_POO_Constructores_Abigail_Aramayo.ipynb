## Ejercicio 1 — Clase Persona
Crear una clase `Persona` con:
- Constructor `__init__` que reciba `nombre`, `edad` y `email` (por defecto `None`)
- Metodo `saludar()` que retorne `"Hola, soy {nombre}"`
- Metodo `es_mayor()` que retorne `True` si la edad es >= 18
- Metodo `cumplir_anios()` que incremente la edad en 1
- Metodo `__str__` que retorne `"Persona({nombre}, {edad})"`

**Pruebas:**
- Crear `p1 = Persona("Ana", 25)` y `p2 = Persona("Carlos", 30, "carlos@mail.com")`
- Imprimir `p1.saludar()` → `"Hola, soy Ana"`
- Verificar `p2.es_mayor()` → `True`
- Ejecutar `p1.cumplir_anios()` y verificar que `p1.edad` sea 26
- Imprimir `p1` → `"Persona(Ana, 26)"`

💡 **Pista:** Una clase se define con `class NombreClase:`. El constructor `__init__(self, ...)` se ejecuta automaticamente al crear un objeto. Usa `self.atributo = valor` para guardar los datos de cada instancia. Recorda que `self` es la referencia al propio objeto.

💡 **Pista:** Un parametro por defecto como `email=None` permite crear objetos sin pasar ese argumento: `Persona("Ana", 25)` funciona porque `email` toma el valor `None` automaticamente.

```python
# Ejercicio 1: Clase Persona

class Persona:
    def __init__(self, nombre, edad, email=None):
        self.nombre = nombre
        self.edad = edad
        self.email = email
    def saludar(self):
        return f"Hola, soy {self.nombre}"
    def es_mayor(self):
        return self.edad >= 18
    def cumplir_años(self):
        self.edad += 1
    def __str__(self):
        return f"Persona({self.nombre}, {self.edad})"

p1 = Persona("Ana", 25)
p2 = Persona("Carlos", 30, "carlos@mail.com")
print(p1.saludar())
print(p2.es_mayor())
p1.cumplir_años ()
print(p1.edad)
print(p1)
```

## Ejercicio 2 — Clase Producto
Crear una clase `Producto` con:
- Constructor que reciba `nombre`, `precio` y `stock` (por defecto `0`)
- Metodo `aplicar_descuento(porcentaje)` que reduzca el precio y lo retorne
- Metodo `hay_stock()` que retorne `True` si el stock es mayor a 0
- Metodo `vender(cantidad=1)` que reduzca el stock si hay suficiente (retorna `True`) o retorne `False` si no alcanza
- Metodo `__str__` que retorne `"{nombre} ${precio} ({stock}u)"`

**Pruebas:**
- Crear `nb = Producto("Notebook", 850000, 15)`
- Aplicar descuento del 10% y verificar nuevo precio
- Vender 3 unidades y verificar stock
- Crear `tk = Producto("Teclado", 45000, 0)` e intentar vender → `False`

💡 **Pista:** Para aplicar un descuento del 10%, multiplica el precio por `(1 - porcentaje / 100)`. El metodo `vender()` debe verificar con un `if` que `cantidad <= self.stock` antes de restar.

```python
# Ejercicio 2: Clase Producto
class Producto:
    def __init__(self, nombre, precio, stock=0):
        self.nombre = nombre
        self.precio = precio
        self.stock = stock
    def aplicar_descuento(self, porcentaje):
        self.precio = self.precio * (1 - porcentaje / 100)
        return self.precio
    def hay_stock(self):
        return self.stock > 0
    def vender(self, cantidad=1):
        if cantidad <= self.stock:
            self.stock -= cantidad
            return True
        else:
            return False
    def __str__(self):
        return f"{self.nombre} ${self.precio} ({self.stock}u)"

nb = Producto("Notebook", 850000, 15)
nb.aplicar_descuento(10)
print(f"Precio con descuento: {nb.precio}")
fue_exitosa = nb.vender(3)
print(f"Venta exitosa: {fue_exitosa}, Stock restante: {nb.stock}")
tk = Producto("Teclado", 45000, 0)
intento_venta = tk.vender()
print(f"Venta de teclado exitosa: {intento_venta}")
print(nb)
print(tk)
```

## Ejercicio 3 — Clase Auto
Crear una clase `Auto` con:
- Constructor que reciba `marca`, `modelo` y `km` (por defecto `0`)
- Atributo `encendido` que inicie en `False`
- Metodo `encender()` que cambie `encendido` a `True`
- Metodo `apagar()` que cambie `encendido` a `False`
- Metodo `conducir(distancia)` que sume la distancia a `km` **solo si el auto esta encendido**
- Metodo `__str__` que retorne `"{marca} {modelo} ({km}km)"`

**Pruebas:**
- Crear `mi_auto = Auto("Toyota", "Corolla")`
- Intentar conducir 100km sin encender → los km deben seguir en 0
- Encender, conducir 150km, verificar km = 150
- Apagar, intentar conducir 50km, verificar que km sigue en 150

```python
# Ejercicio 3: Clase Auto

class Auto:
    def __init__(self, marca, modelo, km=0):
        self.marca = marca
        self.modelo = modelo
        self.km = km
        self.encendido = False
    def encender(self):
        self.encendido = True
    def apagar(self):
        self.encendido = False
    def conducir(self, distancia):
        if self.encendido:
            self.km += distancia
    def __str__(self):
        return f"{self.marca} {self.modelo} ({self.km}km)"

mi_auto = Auto("Toyota", "Corolla")
mi_auto.conducir(100)
print(f"Cantidad de KM despues de intentar conducir sin encender: {mi_auto.km}")
mi_auto.encender()
mi_auto.conducir(150)
print(f"Cantidad de KM al conducir encendido:{mi_auto.km}")
mi_auto.apagar()
mi_auto.conducir(50)
print(f"Cantidad de KM al intentar conducir apagado:{mi_auto.km}")
print(mi_auto)
```

## Ejercicio 4 — Clase CuentaBancaria
Crear una clase `CuentaBancaria` con:
- Constructor que reciba `titular` y `saldo` (por defecto `0`)
- Metodo `depositar(monto)` que sume al saldo **solo si el monto es positivo**
- Metodo `extraer(monto)` que reste del saldo **solo si el monto es positivo y no supera el saldo**. Retorna `True` si se pudo extraer, `False` si no
- Metodo `consultar_saldo()` que retorne el saldo actual
- Metodo `__str__` que retorne `"{titular}: ${saldo}"`

**Pruebas:**
- Crear `cuenta = CuentaBancaria("Ana", 10000)`
- Depositar 5000 → saldo debe ser 15000
- Extraer 3000 → debe retornar `True`, saldo 12000
- Intentar extraer 50000 → debe retornar `False`, saldo sigue en 12000
- Intentar depositar -100 → el saldo no debe cambiar

```python
# Ejercicio 4: Clase CuentaBancaria

class CuentaBancaria:
    def __init__(self, titular, saldo=0):
        self.titular = titular
        self.saldo = saldo
    def depositar(self, monto):
        if monto > 0:
            self.saldo += monto
    def extraer(self, monto):
        if monto > 0 and monto <= self.saldo:
            self.saldo -= monto
            return True
        else:
            return False
    def consultar_saldo(self):
        return self.saldo
    def __str__(self):
        return f"{self.titular}: ${self.saldo}"
cuenta = CuentaBancaria("Ana", 10000)
cuenta.depositar(5000)
print(f"Saldo despues del depósito: {cuenta.consultar_saldo()}")
pudo_extraer = cuenta.extraer(3000)
print(f"Extracción exitosa: {pudo_extraer}, Saldo: {cuenta.consultar_saldo()}")
pudo_extraer_la_cantidad= cuenta.extraer(50000)
print(f"Extracción exitosa: {pudo_extraer_la_cantidad}, Saldo: {cuenta.consultar_saldo()}")
cuenta.depositar(-100)
print(f"Saldo con depósito negativo: {cuenta.consultar_saldo()}")
print(cuenta)
```

## Ejercicio 5 — Clase Estudiante
Crear una clase `Estudiante` con:
- Constructor que reciba `nombre`, `legajo` y una lista `materias` (por defecto lista vacia)
- Metodo `inscribir(materia)` que agregue la materia a la lista **si no esta ya inscripto**
- Metodo `cantidad_materias()` que retorne cuantas materias cursa
- Metodo `cursa(materia)` que retorne `True` si esta inscripto en esa materia
- Metodo `__str__` que retorne `"{nombre} (Leg: {legajo}) - {n} materias"`

**Pruebas:**
- Crear `est = Estudiante("Juan", 12345)`
- Inscribir en "Programacion II" y "Matematica"
- Verificar `est.cursa("Programacion II")` → `True`
- Intentar inscribir de nuevo en "Programacion II" → no debe duplicarse
- Verificar `est.cantidad_materias()` → 2

**Atencion:** Cuidado con usar una lista como parametro por defecto mutable. Investiga por que `materias=[]` puede causar problemas y como solucionarlo.

```python
# Ejercicio 5: Clase Estudiante

class Estudiante:
    def __init__(self, nombre, legajo, materias=None):
        self.nombre = nombre
        self.legajo = legajo
        if materias is None:
            self.materias = []
        else:
            self.materias = materias
    def inscribir(self, materia):
        if materia not in self.materias:
            self.materias.append(materia)
    def cantidad_materias(self):
        return len(self.materias)
    def cursa(self, materia):
        return materia in self.materias
    def __str__(self):
        n = self.cantidad_materias()
        return f"{self.nombre} (Leg: {self.legajo}) - {n} materias"
est = Estudiante("Juan", 12345)
est.inscribir("Programacion II")
est.inscribir("Matematica")
print(f"cursa Programacion II?: {est.cursa('Programacion II')}")
est.inscribir("Programacion II")
print(f"Cantidad de materias: {est.cantidad_materias()}")

print(est)
```

## Ejercicio 6 — Integrador: Concesionaria de Autos
Crear un sistema con dos clases: `Auto` (reusar del Ejercicio 3) y `Concesionaria`.

La clase `Concesionaria` debe tener:
- Constructor que reciba `nombre` e inicialice una lista vacia de autos
- Metodo `agregar_auto(auto)` que agregue un objeto Auto a la lista
- Metodo `buscar_por_marca(marca)` que retorne una lista de autos de esa marca
- Metodo `auto_con_mas_km()` que retorne el auto con mas kilometros
- Metodo `cantidad_autos()` que retorne el total de autos
- Metodo `__str__` que retorne `"Concesionaria {nombre}: {n} autos"`

**Pruebas:**
- Crear una concesionaria y agregar al menos 4 autos de distintas marcas y km
- Buscar por marca "Toyota" y mostrar resultados
- Mostrar el auto con mas km
- Mostrar la concesionaria (usando print)

```python
# Ejercicio 6: Concesionaria de Autos

class Auto:
    def __init__(self, marca, modelo, km=0):
        self.marca = marca
        self.modelo = modelo
        self.km = km
        self.encendido = False
    def encender(self):
        self.encendido = True
    def apagar(self):
        self.encendido = False
    def conducir(self, distancia):
        if self.encendido:
            self.km += distancia
    def __str__(self):
        return f"{self.marca} {self.modelo} ({self.km}km)"
class Concesionaria:
    def __init__(self, nombre):
        self.nombre = nombre
        self.autos = []
    def agregar_auto(self, auto):
        self.autos.append(auto)
    def buscar_por_marca(self, marca):
        lista_filtrada = []
        for auto in self.autos:
            if auto.marca == marca:
                lista_filtrada.append(auto)
        return lista_filtrada
    def auto_con_mas_km(self):
        if not self.autos:
            return None
        mas_km = self.autos[0]
        for auto in self.autos:
            if auto.km > mas_km.km:
                mas_km = auto
        return mas_km

    def cantidad_autos(self):
        return len(self.autos)

    def __str__(self):
        return f"Concesionaria {self.nombre}: {self.cantidad_autos()} autos"
mi_agencia = Concesionaria("AAA")
mi_agencia.agregar_auto(Auto("Toyota", "Corolla", 15000))
mi_agencia.agregar_auto(Auto("Ford", "Eco Sport", 95000))
mi_agencia.agregar_auto(Auto("Toyota", "Prius", 50000))
mi_agencia.agregar_auto(Auto("Fiat", "Cronos", 33000))
print("Autos marca Toyota:")
toyotas = mi_agencia.buscar_por_marca("Toyota")
for t in toyotas:
    print(t)
print(f"Auto con más kilometraje: {mi_agencia.auto_con_mas_km()}")

print(mi_agencia)
```
